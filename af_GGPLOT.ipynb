```r
male_players <- read_csv("data/male_players.csv")
```

```r
male_players
```

```r
male_players |> Filter(age >= 18 & height_cm >= 100 & weight_kg >= 50)
```

```r
male_players
```

```r
male_players |> select(age, height = height_cm, weight = weight_kg) |> filter(age >= 18 & height >= 100 & weight >= 50) -> demographs
```

```r
demographs
```

```r
demographs |> ggplot() +aes(x = age, y = height) + geom_point() + geom_smooth(method=lm) +
  theme_minimal()
```

```r
demographs |> ggplot() +aes(x = height) + geom_histogram()
```

```r
demographs |> ggplot() +aes(x = height) + geom_histogram(color ='red')
```

```r
demographs |> ggplot() +aes(x = height) + geom_histogram(fill = 'blue', color ='white')
```

```r
demographs |> ggplot() +aes(x = weight) + geom_boxplot(fill = 'gold', color ='black')
```
